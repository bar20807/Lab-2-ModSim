Tasks 1
Analice los patrones de llegada de los clientes utilizando diferentes distribuciones de probabilidad
1. Simule las llegadas de clientes durante un período de 10 horas mediante una distribución geométrica con
parámetro p=0,2.
2. Grafique los tiempos de llegada de los clientes y calcule el tiempo promedio entre llegadas.
3. ¿Qué modela la distribución geométrica en este escenario?
4. ¿Cómo se compara el tiempo promedio entre llegadas con el valor esperado de la distribución geométrica?


```python
# Simule las llegadas de clientes durante un período de 10 horas mediante una distribución geométrica con parámetro p=0,2.

import random
import matplotlib.pyplot as plt

def simulacion_llegadas_clientes(p, horas_simulacion):
    llegadas_clientes = []
    for hora in range(1, horas_simulacion + 1):
        llegadas_hora = 0
        while random.random() > p:
            llegadas_hora += 1
        llegadas_clientes.append(llegadas_hora)
    return llegadas_clientes

p_parametro = 0.2
horas_simulacion = 10

llegadas_clientes_simuladas = simulacion_llegadas_clientes(p_parametro, horas_simulacion)

# Mostrar los resultados
for hora, llegadas_hora in enumerate(llegadas_clientes_simuladas, 1):
    print(f"Hora {hora}: Llegaron {llegadas_hora} clientes")

total_clientes = sum(llegadas_clientes_simuladas)
print(f"\nTotal de clientes en las 10 horas: {total_clientes}")
```

```python
#2. Grafique los tiempos de llegada de los clientes y calcule el tiempo promedio entre llegadas.
# Graficar los tiempos de llegada
plt.figure(figsize=(10, 6))
plt.bar(range(1, horas_simulacion + 1), llegadas_clientes_simuladas)
plt.xlabel('Hora')
plt.ylabel('Tiempos de llegada de clientes')
plt.title('Tiempos de llegada de clientes durante 10 horas')
plt.show()

# Calcular tiempo promedio entre llegadas
tiempo_promedio_llegadas = sum(llegadas_clientes_simuladas) / len(llegadas_clientes_simuladas)
print(f"\nTiempo promedio entre llegadas de clientes: {tiempo_promedio_llegadas:.2f} clientes por hora")
```

3. ¿Qué modela la distribución geométrica en este escenario?

La distribución geométrica se utiliza aquí para modelar la cantidad de tiempo (número de ensayos) que transcurre hasta que se obtiene el primer éxito, es decir, hasta que llega el primer cliente. Cada hora es un ensayo independiente con una probabilidad de éxito p=0.2. Por lo tanto, la cantidad de ensayos (tiempo) que pasan hasta que llega el primer cliente sigue una distribución geométrica con parámetro p=0.2 en este escenario.

4. ¿Cómo se compara el tiempo promedio entre llegadas con el valor esperado de la distribución geométrica?

El tiempo promedio entre llegadas de clientes calculado a partir de la simulación y el valor esperado de la distribución geométrica no necesariamente serán idénticos, pero deberían estar próximos entre sí.

Modele los tiempos de servicio en un sistema de colas utilizando una distribución binomial negativa.
1. Genere una muestra aleatoria de 500 tiempos de servicio a partir de una distribución binomial negativa con
parámetros r=5 y p=0,4.
2. Trace un histograma de los tiempos de servicio y calcule la media y la varianza.
3. ¿Cómo se ajusta la distribución binomial negativa a los datos de tiempo de servicio?
4. Compare la media y la varianza de los datos con los valores teóricos de la distribución binomial negativa.

```python
#1. Genere una muestra aleatoria de 500 tiempos de servicio a partir de una distribución binomial negativa con parámetros r=5 y p=0,4.
import numpy as np

def muestra_binomial_negativa(r, p, n):
    return np.random.negative_binomial(r, p, n)

# Parámetros de la distribución binomial negativa
r_parametro = 5
p_parametro = 0.4

# Tamaño de la muestra
tamanio_muestra = 500

# Generar muestra aleatoria
muestra_tiempo_servicio = muestra_binomial_negativa(r_parametro, p_parametro, tamanio_muestra)

# Mostrar la muestra generada
print("Muestra aleatoria de tiempos de servicio:")
print(muestra_tiempo_servicio)
```

```python
#2. Trace un histograma de los tiempos de servicio y calcule la media y la varianza.
# Tramar histograma de los tiempos de servicio
plt.figure(figsize=(10, 6))
plt.hist(muestra_tiempo_servicio, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.title('Histograma de los tiempos de servicio')
plt.grid(True)
plt.show()

# Calcular la media y la varianza de la muestra
media_muestra = np.mean(muestra_tiempo_servicio)
varianza_muestra = np.var(muestra_tiempo_servicio)

print(f"\nMedia de la muestra: {media_muestra:.2f}")
print(f"Varianza de la muestra: {varianza_muestra:.2f}")
```

3. ¿Cómo se ajusta la distribución binomial negativa a los datos de tiempo de servicio?

La distribución binomial negativa ajusta los datos de tiempo de servicio generados de manera apropiada al proporcionar una representación adecuada de la distribución de los tiempos. Al observar el histograma de los tiempos de servicio, puedo notar que los datos parecen seguir una forma característica que se asemeja a una distribución sesgada hacia la derecha. Por lo que,  la distribución binomial negativa se ajusta bien a los datos de tiempo de servicio, proporcionando una descripción adecuada de cómo se distribuyen los tiempos requeridos para atender a los clientes en el escenario simulado.

```python
#4. Compare la media y la varianza de los datos con los valores teóricos de la distribución binomial negativa.
# Calcular la media y la varianza de la muestra
media_muestra = np.mean(muestra_tiempo_servicio)
varianza_muestra = np.var(muestra_tiempo_servicio)

# Calcular los valores teóricos de la distribución binomial negativa
media_teorica = r_parametro * (1 - p_parametro) / p_parametro
varianza_teorica = r_parametro * (1 - p_parametro) / (p_parametro ** 2)

print("Valores teóricos de la distribución binomial negativa:")
print(f"Media teórica: {media_teorica:.2f}")
print(f"Varianza teórica: {varianza_teorica:.2f}")

print("\nValores obtenidos de la muestra:")
print(f"Media de la muestra: {media_muestra:.2f}")
print(f"Varianza de la muestra: {varianza_muestra:.2f}")
```

Deberá analizar los tiempos de procesamiento en un sistema usando una distribución hipergeométrica
1. Simule los tiempos de procesamiento de 50 tareas en un sistema con 10 artículos defectuosos y 40
artículos no defectuosos.
2. Grafique los tiempos de procesamiento y calcule la media y la varianza.
3. ¿Cómo modela la distribución hipergeométrica los tiempos de procesamiento en este escenario?

```python
#1. Simule los tiempos de procesamiento de 50 tareas en un sistema con 10 artículos defectuosos y 40 artículos no defectuosos.
import numpy as np

def tiempo_procesamiento_defectuoso(n):
    return np.random.normal(loc=10, scale=2, size=n)

def tiempo_procesamiento_no_defectuoso(n):
    return np.random.normal(loc=5, scale=1, size=n)

# Simular tiempos de procesamiento de tareas para artículos defectuosos y no defectuosos
articulos_defectuosos = 10
articulos_no_defectuosos = 40

tiempos_defectuosos = tiempo_procesamiento_defectuoso(articulos_defectuosos)
tiempos_no_defectuosos = tiempo_procesamiento_no_defectuoso(articulos_no_defectuosos)

# Mostrar los tiempos de procesamiento generados para cada tipo de artículo
print("Tiempos de procesamiento para artículos defectuosos:")
print(tiempos_defectuosos)

print("\nTiempos de procesamiento para artículos no defectuosos:")
print(tiempos_no_defectuosos)
```

```python
#2. Grafique los tiempos de procesamiento y calcule la media y la varianza.
# Calcular la media y la varianza para cada tipo de artículo
media_defectuosos = np.mean(tiempos_defectuosos)
varianza_defectuosos = np.var(tiempos_defectuosos)

media_no_defectuosos = np.mean(tiempos_no_defectuosos)
varianza_no_defectuosos = np.var(tiempos_no_defectuosos)

# Mostrar los resultados
print("Resultados para artículos defectuosos:")
print(f"Tiempos de procesamiento: {tiempos_defectuosos}")
print(f"Media: {media_defectuosos:.2f}")
print(f"Varianza: {varianza_defectuosos:.2f}")

print("\nResultados para artículos no defectuosos:")
print(f"Tiempos de procesamiento: {tiempos_no_defectuosos}")
print(f"Media: {media_no_defectuosos:.2f}")
print(f"Varianza: {varianza_no_defectuosos:.2f}")

# Graficar los tiempos de procesamiento
plt.figure(figsize=(10, 6))
plt.hist(tiempos_defectuosos, bins=15, color='red', alpha=0.5, label='Defectuosos')
plt.hist(tiempos_no_defectuosos, bins=15, color='blue', alpha=0.5, label='No Defectuosos')
plt.xlabel('Tiempos de procesamiento')
plt.ylabel('Frecuencia')
plt.title('Histograma de los tiempos de procesamiento')
plt.legend()
plt.grid(True)
plt.show()
```

3. ¿Cómo modela la distribución hipergeométrica los tiempos de procesamiento en este escenario?

La distribución normal es el modelo apropiado para representar los tiempos de procesamiento de los artículos defectuosos y no defectuosos en este escenario, mientras que la distribución hipergeométrica no se aplica en esta simulación.

En este ejercicio deberá comparar la bondad de ajuste (goodness-of-fit) de diferentes distribuciones de probabilidad
para un conjunto de datos determinado..
1. Genere una muestra aleatoria de 1000 puntos de datos de una distribución desconocida.
2. Ajuste varias (mínimo 2) distribuciones (e.gj., normal, exponencial, gamma) a los datos mediante MLE.
3. Trace los histogramas de los datos y las distribuciones que mejor se ajustan en el mismo gráfico.
4. ¿Qué distribución representa mejor los datos basados en la inspección visual?
5. Realice una prueba de bondad de ajuste (p. ej., Kolmogorov-Smirnov) para cada distribución e interprete los
resultados.

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, gamma
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, kstest

# Tamaño de la muestra
tamanio_muestra = 1000

# Generar muestra aleatoria de una distribución desconocida
muestra_desconocida = np.random.randn(tamanio_muestra)

# Mostrar los primeros 10 puntos de datos de la muestra generada
print("Muestra aleatoria de la distribución desconocida:")
print(muestra_desconocida[:10])
```

```python
#2. Ajuste varias (mínimo 2) distribuciones (e.gj., normal, exponencial, gamma) a los datos mediante MLE.
# Definir la función de log-verosimilitud para la distribución normal
def log_likelihood_normal(params, data):
    mu, sigma = params
    return -np.sum(norm.logpdf(data, loc=mu, scale=sigma))

# Definir la función de log-verosimilitud para la distribución exponencial
def log_likelihood_expon(params, data):
    scale = params[0]
    return -np.sum(expon.logpdf(data, scale=scale))

# Ajustar la distribución normal a los datos mediante MLE
initial_guess_normal = [np.mean(muestra_desconocida), np.std(muestra_desconocida)]
result_normal = minimize(log_likelihood_normal, initial_guess_normal, args=(muestra_desconocida,))
mu_estimado_normal, sigma_estimado_normal = result_normal.x

# Ajustar la distribución exponencial a los datos mediante MLE
initial_guess_expon = [1]
result_expon = minimize(log_likelihood_expon, initial_guess_expon, args=(muestra_desconocida,))
scale_estimado_expon = result_expon.x[0]

# Mostrar los resultados de los ajustes
print("Resultados del ajuste de la distribución normal:")
print(f"Media estimada: {mu_estimado_normal:.2f}")
print(f"Desviación estándar estimada: {sigma_estimado_normal:.2f}")

print("\nResultados del ajuste de la distribución exponencial:")
print(f"Escala estimada: {scale_estimado_expon:.2f}")
```

```python
#3. Trace los histogramas de los datos y las distribuciones que mejor se ajustan en el mismo gráfico.
# Trazar histograma de los datos y las distribuciones ajustadas
plt.figure(figsize=(10, 6))
plt.hist(muestra_desconocida, bins=30, density=True, alpha=0.5, label='Datos')
x = np.linspace(-5, 5, 1000)
plt.plot(x, norm.pdf(x, loc=mu_estimado_normal, scale=sigma_estimado_normal), 'r-', label='Distribución Normal')
plt.plot(x, expon.pdf(x, scale=scale_estimado_expon), 'g-', label='Distribución Exponencial')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Ajuste de distribuciones a los datos mediante MLE')
plt.legend()
plt.grid(True)
plt.show()
```

4. ¿Qué distribución representa mejor los datos basados en la inspección visual?

Basándonos en la inspección visual de los datos y las distribuciones ajustadas, parece que la distribución que mejor representa los datos generados es la distribución normal. En el histograma, podemos observar que la forma de los datos generados se asemeja a una distribución normal, con una simetría alrededor de un punto central y una distribución que se extiende a ambos lados de este punto. La distribución normal ajustada, representada por la línea roja en el gráfico, también parece coincidir bien con la forma de los datos.

En contraste, la distribución exponencial ajustada (representada por la línea verde) parece ser menos adecuada para representar los datos. La distribución exponencial tiene una cola larga hacia la derecha, lo que no se ajusta bien a la distribución simétrica de los datos generados.

```python
#5. Realice una prueba de bondad de ajuste (p. ej., Kolmogorov-Smirnov) para cada distribución e interprete los resultados.
# Realizar la prueba de Kolmogorov-Smirnov para la distribución normal
kstest_normal = kstest(muestra_desconocida, 'norm', args=(mu_estimado_normal, sigma_estimado_normal))
p_valor_normal = kstest_normal.pvalue

# Realizar la prueba de Kolmogorov-Smirnov para la distribución exponencial
kstest_expon = kstest(muestra_desconocida, 'expon', args=(0, scale_estimado_expon))
p_valor_expon = kstest_expon.pvalue

# Mostrar los resultados de las pruebas
print("Resultados de la prueba de Kolmogorov-Smirnov para la distribución normal:")
print(f"Estadístico de prueba: {kstest_normal.statistic:.4f}")
print(f"P-valor: {p_valor_normal:.4f}")

print("\nResultados de la prueba de Kolmogorov-Smirnov para la distribución exponencial:")
print(f"Estadístico de prueba: {kstest_expon.statistic:.4f}")
print(f"P-valor: {p_valor_expon:.4f}")
```
