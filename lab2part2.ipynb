1. ¿Puede ser este evento modelado por una distribución de Poisson? ¿Por qué?

Sí, este evento puede ser modelado por una distribución de Poisson. La distribución de Poisson es adecuada para describir la cantidad de eventos que ocurren en un intervalo de tiempo o espacio específico cuando estos eventos ocurren de forma independiente y a una tasa constante.

En este caso, estamos considerando la cantidad promedio de buses que llegan a una parada de bus cada 30 minutos, con un promedio de 2 buses por intervalo. Suponemos que la llegada de buses es independiente y que la tasa promedio de llegada es constante en todos los intervalos.

Al cumplirse estas condiciones, podemos utilizar la distribución de Poisson para calcular la probabilidad de que llegue un número específico de buses en un período de 30 minutos. Esta distribución nos permite modelar adecuadamente la variabilidad en la llegada de buses a la parada, y nos ayuda a comprender mejor las probabilidades asociadas con diferentes cantidades de buses en un intervalo de tiempo dado.

```python
#2. Calcule y grafique la probabilidad para diferentes números de buses, yendo desde 0 hasta 100. ¿Cuál es la cantidad de buses más probable?

import math
import matplotlib.pyplot as plt

def poisson_probability(lambda_value, k):
    return (math.exp(-lambda_value) * (lambda_value ** k)) / math.factorial(k)

def plot_poisson_distribution(lambda_value, max_k):
    buses = list(range(max_k + 1))
    probabilities = [poisson_probability(lambda_value, k) for k in buses]
    
    plt.bar(buses, probabilities)
    plt.xlabel('Número de Buses')
    plt.ylabel('Probabilidad')
    plt.title('Distribución de Poisson para llegada de Buses')
    plt.show()

lambda_value = 2
max_buses = 100
plot_poisson_distribution(lambda_value, max_buses)
```

```python
#1. Genere una tabla que muestre los tiempos intermedios en horas para los 10 primeros pacientes
import numpy as np

# Tasa promedio de llegada de pacientes por hora (lambda)
tasa_llegada = 5

# Número de pacientes a generar
num_pacientes = 10

# Generar tiempos intermedios utilizando la inversa de la CDF exponencial
tiempos_intermedios = np.random.exponential(scale=1/tasa_llegada, size=num_pacientes)

# Crear la tabla con los tiempos intermedios
tabla_tiempos_intermedios = []
for i, tiempo in enumerate(tiempos_intermedios, 1):
    tabla_tiempos_intermedios.append((i, tiempo))

# Mostrar la tabla de tiempos intermedios
print("Tabla de tiempos intermedios para los primeros 10 pacientes:")
print("--------------------------------------------------------")
print("Paciente | Tiempo intermedio (horas)")
print("-----------------------------------")
for paciente, tiempo in tabla_tiempos_intermedios:
    print(f"{paciente:8} | {tiempo:.2f}")
```

```python
#2. Grafique usando el mismo lambda la CDF para la variable exponencial.
# Definir el rango de valores para x (tiempos intermedios)
x = np.linspace(0, 5, 100)  # Valores de 0 a 5 horas

# Calcular la CDF para cada valor de x utilizando la fórmula de la distribución exponencial
cdf = 1 - np.exp(-tasa_llegada * x)

# Graficar la CDF
plt.plot(x, cdf, label='CDF')
plt.xlabel('Tiempo intermedio (horas)')
plt.ylabel('Probabilidad acumulada')
plt.title('CDF de la distribución exponencial')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#3. Haga una gráfica de los tiempos intermedios para los primeros 500 pacientes. ¿Qué forma tiene la gráfica?
# Número de pacientes a generar
num_pacientes = 500

# Generar tiempos intermedios utilizando la inversa de la CDF exponencial
tiempos_intermedios = np.random.exponential(scale=1/tasa_llegada, size=num_pacientes)

# Crear una lista con los números de paciente (del 1 al 500)
numeros_pacientes = np.arange(1, num_pacientes + 1)

# Graficar los tiempos intermedios
plt.plot(numeros_pacientes, tiempos_intermedios)
plt.xlabel('Número de paciente')
plt.ylabel('Tiempo intermedio (horas)')
plt.title('Tiempos intermedios para los primeros 500 pacientes')
plt.grid(True)
plt.show()
```

Al observar la gráfica de los tiempos intermedios para los primeros 500 pacientes, puedo notar que tiene una forma que se asemeja a una línea ascendente con cierta variabilidad. A medida que aumenta el número de paciente, también aumenta el tiempo intermedio entre llegadas. Sin embargo, debido a la naturaleza aleatoria de la distribución exponencial, los tiempos intermedios no siguen un patrón lineal perfecto, sino que muestran cierta variación alrededor de la tendencia ascendente. Esta gráfica es característica de un proceso de Poisson, donde los intervalos de tiempo entre eventos consecutivos siguen una distribución exponencial.

1. ¿Son las variables T1, T2 hasta Tk variables aleatorias? ¿Por qué?


Sí, las variables T1, T2, ..., Tk son variables aleatorias. Esto se debe a que los tiempos intermedios X1, X2, ..., Xk son variables aleatorias que representan los intervalos de tiempo entre llegadas consecutivas de pacientes a la clínica veterinaria. Al sumar estos tiempos intermedios para obtener las variables T1, T2, ..., Tk, estamos generando el tiempo de llegada del primer, segundo, tercer, hasta el k-ésimo paciente.

Dado que los tiempos intermedios son variables aleatorias que siguen una distribución exponencial, las variables T1, T2, ..., Tk también son variables aleatorias. La suma de variables aleatorias sigue siendo una variable aleatoria. Por lo tanto, el tiempo de llegada de cada paciente a la clínica veterinaria es una variable aleatoria, ya que depende de la suma de múltiples tiempos intermedios, que a su vez son generados aleatoriamente según la distribución exponencial.

```python
#2. Simule y grafique el proceso de Poisson completo usando el mismo gamma e información recabada del ejercicio 2. Haga una simulación para 100 pacientes.

import numpy as np
import matplotlib.pyplot as plt

# Tasa promedio de llegada de pacientes por hora (lambda)
tasa_llegada = 5

# Número de pacientes a simular
num_pacientes = 100

# Generar tiempos intermedios utilizando la inversa de la CDF exponencial
tiempos_intermedios = np.random.exponential(scale=1/tasa_llegada, size=num_pacientes)

# Calcular los tiempos de llegada acumulados (T1, T2, ..., Tk)
tiempos_llegada = np.cumsum(tiempos_intermedios)

# Crear una lista con los números de paciente (del 1 al 100)
numeros_pacientes = np.arange(1, num_pacientes + 1)

# Graficar los tiempos de llegada acumulados (proceso de Poisson)
plt.plot(numeros_pacientes, tiempos_llegada)
plt.xlabel('Número de paciente')
plt.ylabel('Tiempo de llegada (horas)')
plt.title('Proceso de Poisson para 100 pacientes')
plt.grid(True)
plt.show()
```
